KDE - Kernel Density Estimator

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
import matplotlib.pyplot as plt

sns.displot(df['total_bill'])
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

```python
sns.displot(df['total_bill'],kde=False,bins=30);
```

jointplot
jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what kind parameter to compare with:

“scatter”
“reg”
“resid”
“kde”
“hex”

```python
sns.jointplot(x="total_bill",y='tip',data=df,kind='kde');
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(df['total_bill'])
```

```python
sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```

```python
sns.kdeplot(df['tip'])
sns.rugplot(df['tip'])
```

Categorical Data Plots
Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

factorplot
boxplot
violinplot
stripplot
swarmplot
barplot
countplot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=df,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=df, palette="coolwarm")
```

```python
sns.violinplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=df,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=df,hue='sex',split=True,palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

```python
sns.stripplot(x="day", y="total_bill", data=df,jitter=True,hue='sex',palette='Set1')
```
