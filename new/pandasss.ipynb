```python
import pandas as pd
```

```python
df = pd.read_excel("EXCEL_catalog.xlsx")
```

```python
column_mapping =  {
    "ref":"CODE",
    "categorie":"TYPE",
    "title":"HEADING",
    "tags":"LABELS",
    "price":"COST",
    "description":"DETAILS"
}
```

```python
df = df.rename(columns=column_mapping)
```

```python
columns = ['CODE', 'TYPE', 'HEADING', 'LABELS', 'COST', 'DETAILS']
```

```python
# Reorder the DataFrame using the new column order
df = df[columns]
```

```python
df.to_csv("CSV_catalog.csv", index=False)

print(f"Data has been successfully imported")
```

```python
# Remove Special Characters
df["CODE"] = df["CODE"].str.strip("Â")
print("The Special Characters was removed")
```

```python
# Add Currency Symbol
df['COST'] = df['COST'].apply(lambda x: f'{x:.2f} INR')
print("The currency symbol was added")
```

```python
default_costs = {
    'earrings': '5000 INR',
    'necklaces': '6000 INR',
    'rings': '7000 INR',
    'bracelets': '8000 INR'
}

def fill_cost(row):
    if pd.isna(row['COST']) or 'nan' in str(row['COST']):
        item_type = row['TYPE']
        return default_costs.get(item_type, row['COST'])  
    return row['COST']

# Apply the function to the DataFrame
df['COST'] = df.apply(fill_cost, axis=1)
print("Missing cost values were filled")
```

```python
df
```

```python
stock_mapping = {
    'earrings': 15,
    'necklaces': 20,
    'rings': 25,
    'bracelets': 10
}
df['STOCKS'] = df['TYPE'].replace(stock_mapping)
```

```python
df.to_csv("CSV_catalog.csv", index=False)
```

```python
duplicates = df[df.duplicated(keep=False)]
duplicates.to_csv("Duplicate.csv", index=False)

print("Duplicate entries have been successfully saved to 'Duplicate.csv'.")
```

```python
# Remove Duplicate Entries
df = df.drop_duplicates()
print("The Duplicates are removed")
df.to_csv("CSV_catalog.csv", index=False)
```

```python
# Remove 'INR' and convert to float
df['COST'] = df['COST'].replace({' INR': '', ',': ''}, regex=True).astype(float)
df = df.sort_values(by=['TYPE', 'COST'], ascending=[True, True])
df['COST'] = df['COST'].apply(lambda x: f'{x:.2f} INR')
print("The currency symbol was added")
df.to_csv("CSV_catalog.csv", index=False)
```
