```python
import numpy as np
import pandas as pd

df = pd.read_csv('survey lung cancer.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['LUNG_CANCER']=encoder.fit_transform(df['LUNG_CANCER'])
df['GENDER']=encoder.fit_transform(df['GENDER'])
df.head()
```

```python
df.corr()
```

```python
x=df.drop(['LUNG_CANCER'],axis=1)
y=df['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3, random_state=42 )
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

clf =  DecisionTreeClassifier( random_state = 40)
clf.fit(x_train,  y_train)
```

```python
clf.get_depth()
```

```python
clf.get_n_leaves()
```

```python
pred = clf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
tpred = clf.predict(x_train)
from sklearn.metrics import accuracy_score
accuracy_score(y_train, tpred)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize = (15, 15))
tree.plot_tree(clf, filled = True)
plt.show()
```

```python
path = clf.cost_complexity_pruning_path(x_train,  y_train)
path 
```

```python
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
ccp_alphas
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf =  DecisionTreeClassifier(random_state = 40, ccp_alpha=ccp_alpha)  
    clf.fit(x_train,  y_train)
    print('Alpha',  ccp_alpha)
    print('Node_count', clf.tree_.node_count)
    clfs.append(clf)
print("No of node in last tree is: {} with ccp_alpha: {}".format(clfs[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
trains_scores = [clf.score(x_train, y_train) for clf in clfs]
test_scores = [clf.score(x_test, y_test) for clf in clfs]
```
