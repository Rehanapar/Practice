```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 23)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = pd.DataFrame(scaler.fit_transform(x_train))
x_test = pd.DataFrame(scaler.transform(x_test)) 
```

```python
col = ['day', 'pressure ', 'maxtemp', 'temparature', 'mintemp', 'dewpoint',
       'humidity ', 'cloud ', 'sunshine', 'winddirection',
       'windspeed']
```

```python
model = ExtraTreesClassifier()
model.fit(x_train, y_train)
```

```python
imp = pd.Series(model.feature_importances_, index = col)
imp
```

```python
imp.nlargest(5).plot(kind = 'bar')
```

```python
x = imp.nlargest(5)
x
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix 
def Metrics(model, x, y):
    y_pred = model.predict(x)
    ac_sc = accuracy_score(y, y_pred)
    print(ac_sc)

    pr_sc = precision_score(y, y_pred)
    print(pr_sc)


Metrics(model, x_train, y_train)
```
