```python
import spacy
from nltk.corpus import stopwords
import nltk
import pandas as pd
import re
import string
 
# Download NLTK stopwords
nltk.download('stopwords')
 
# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
 
# Get NLTK stopwords and add custom ones
nltk_stopwords = set(stopwords.words('english') + ['u', '', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure'])
```

```python
 
def preprocess_tweet(tweet):
    tweet = str(tweet)
    tweet = tweet.lower()
    
    # Remove URLs
    tweet = re.sub(r'https?://(www\.)?[a-z0-9-]+(\.[a-z0-9-]+)+(\/[a-z0-9-]+)*\/?', '', tweet)
    
    # Remove Mentions
    tweet = re.sub(r'@[A-Za-z0-9_]*', '', tweet)
    
    # Remove hashtags
    tweet = re.sub('#', '', tweet)
    
    # Remove punctuation
    tweet = ''.join([char for char in tweet if char not in string.punctuation])
    
    # Process the tweet using spaCy
    doc = nlp(tweet)
    
    # Remove stopwords
    filtered_tokens = [token.text for token in doc if token.text not in nltk_stopwords]
    
    return " ".join(filtered_tokens)
 
# # # Example DataFrames
# # train_df = pd.DataFrame({'text': ["This is a sample tweet! Check http://example.com #example @user"]})
# # test_df = pd.DataFrame({'text': ["Another example tweet! Visit https://example.org #hashtag @mention"]})
 
```

```python
# Apply the preprocessing function
train_df['preprocessed_text'] = train_df['text'].apply(preprocess_tweet)
test_df['preprocessed_text'] = test_df['text'].apply(preprocess_tweet)
 
# Display the results
print(train_df)
print(test_df)
```
