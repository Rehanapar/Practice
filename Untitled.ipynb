```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.datasets import fetch_california_housing
import numpy as np

# Load dataset (replace this with your own dataset if necessary)
data = fetch_california_housing()
features = data.data  # This should be the actual data (2D array)
target = data.target

# Initialize PolynomialFeatures
interaction = PolynomialFeatures(degree=3, include_bias=False, interaction_only=True)

# Fit and transform the features (ensure it's 2D)
features_interaction = interaction.fit_transform(features)

# Check the transformed features
print(features_interaction)
```

```python
from sklearn.linear_model import LinearRegression
regression =  LinearRegression()
model = regression.fit(features_interaction, target)
```
